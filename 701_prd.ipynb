```python
P_=np.dot(w_matrix, w_matrix.T)
```

```python
P_arr = np.array([P_ for _ in range(30)])
```

```python
''' 定义状态转移协方差矩阵Q 这里我们把协方差设置的很小，因为觉得状态转移矩阵准确度高 '''
global Q_
Q_=0.0001*np.eye(32)
Q_
```

```python
from cvxopt import matrix, solvers
import cvxopt
```

```python
R = np.mat([30])
w_matrix_post =  w_matrix.T
w_matrix_predict = w_matrix.T
w_prior = np.zeros([30,32])
P_predict = P_arr
P_post = np.zeros([30,32,32])

w_matrix_prior_next=np.zeros([30,32])

P_prior_next=np.zeros([30,32,32])

#for i in range(30):
#    P_predict[i] = P_arr[i]
for i in range(30):
    K_ = np.matrix(P_predict[i])*r_matrix.T / (r_matrix * np.matrix(P_predict[i]) * r_matrix.T + R)
    #w_matrix_[:,i]
    #每个基金未优化权重
    w_unnorm = w_matrix_predict[i].T + K_ *(R_matrix[0,i] - r_matrix * w_matrix_predict[i].T)

    #youhua
    q_mat = matrix((-w_unnorm).A1.tolist(),(32,1))
    P_mat = matrix(np.eye(32))
    #P_mat = matrix(P_)
    A = np.ones(32)

    A = matrix(A).T
    b = matrix([1.0])
    G = matrix(-np.eye(32))
    h = np.zeros(32)
    h_column = h.reshape(-1, 1)
    h = matrix(h)
    cvxopt.solvers.options['show_progress'] = False

    result = solvers.qp(P_mat,q_mat,G,h,A,b)

    ws = np.round(np.array(result['x']),4)
    #print(ws)



    w_matrix_post[i]=ws.T
    #print(w_matrix_post)


    P_post[i] = (np.eye(32) - K_ * r_matrix) * P_predict[i]

    #dui P youhua
    zero_indices = np.where(ws == 0)[0]
    if  len(zero_indices)==0:
      w_matrix_prior_next[i]=w_matrix_post[i]

      P_prior_next[i]=P_post[i]+Q_
      continue


    else:

      D_new = np.zeros(32)
      D_new[zero_indices] = 1
      D_new = np.matrix(D_new)
      tao = D_new.T/(D_new*D_new.T)
      P_new = (np.matrix(np.eye(32))-tao*D_new)*P_arr[i]*(np.matrix(np.eye(32))-tao*D_new).T

      P_post[i] = P_new

    #propagation

    w_matrix_prior_next[i]=w_matrix_post[i]

    P_prior_next[i]=P_post[i]+Q_
```
